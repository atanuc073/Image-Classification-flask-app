```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import skimage.io
import skimage.feature
import skimage.transform

import os
from glob import glob
import re
%matplotlib inline
```

```python
# Read all the folders in a list
folder=os.listdir("data")
print(folder)
```

```python
all_path=[]
for f in folder:
    all_path +=glob('./data/{}/*.jpg'.format(f))
```

```python
len(all_path)
```

```python
## Read all images and Label
img_all=skimage.io.ImageCollection(all_path)
len(img_all)
```

```python
img_all.files
```

```python
# Labeling
# extract text Folder
def extract_label(text):
    try:
        label=re.search(r'./data/(.*?)Head/',text).group(1)
    except:
        label=re.search(r'./data/(.*?)/',text).group(1)
    return label.lower()
```

```python
labels=list(map(extract_label,img_all.files))
```

```python
len(labels)
```

```python
img_all[0]
```

```python
# load all the images
def buffer(io):
    return io
img_all_arrs=np.array(list(map(buffer,img_all)))
```

```python
img_all_arrs.shape
```

```python

data=dict()
data["description"]="There are 20 classes and 2023 images, The images are 80 X 80 X (rgb)"
data["data"] = img_all_arrs
data["target"] = labels
data["labels"] = set(labels)
```

```python
# save the file in pickle
# import pickle
# import collections
# pickle.dump(data,open('data_animals_head_20.pickle','wb'))
```

```python
# How many pic for each class label
import collections
count_values=collections.Counter(data['target'])
count_values
```

```python
y_axis= list(count_values.keys())
x_axis=count_values.values()
plt.figure(figsize=(10,6))
plt.title("Class Count")
plt.xlabel("Count")
plt.barh(y_axis,x_axis);
```

```python
plt.imshow(data["data"][210])
plt.xticks([]),plt.yticks([])
plt.title(data['target'][210])
plt.show()
```

```python
## See all class Images
plt.figure(figsize=(14,8))
for i,c in enumerate(data["labels"]):
    index=data["target"].index(c)
    img=data["data"][index]
    plt.subplot(3,7,i+1)
    plt.imshow(data["data"][index])
    plt.xticks([]),plt.yticks([])
    plt.title(c)
plt.show()
```
